# Target detection
## We need to compare the image types by computing mean, variance, skewness and kurtosis for each image type and then comparing the same.
### We can start by any one image type and follow the same procedure for other image types.

# 1. Human analysis
   ## Following are the steps:
   ## 1. Import the images and let all the images have 300*300 pixels.
   ## 2. Compute the grey scale value of each and every image.

```python
from scipy.misc import imread, imsave, imresize
import matplotlib.pyplot as plt

import numpy as np
man_g=[np.uint8(np.zeros((300,300))) for i in range(16)]
%matplotlib inline
man = [imresize(imread('C:/Users/sunny/Desktop/data/4/'+str(i+1)+'.jpg'),(300,300)) for i in range(16)]
#plt.imshow(img[0])
for k in range(16):
    for i in range(298):
        for j in range(298):
            man_g[k][i+1,j+1]=np.uint8(((np.float32(man[k][i,j,0])+np.float32(man[k][i,j,1])+np.float32(man[k][i,j,2]))/3))
            

        
```

## 3. After computing the grey scale, compute the mean of all the images.

```python
import numpy as np
man_m=[np.uint8(np.zeros((300,300)))for i in range(16)]
for k in range(16):
    for i in range(298):
        for j in range(298):
            
            man_m[k][i+1,j+1]= np.uint8(((np.float32(man_g[k][i,j])+np.float32(man_g[k][i,j+1])+np.float32(man_g[k][i,j+2])+
                                    np.float32(man_g[k][i+1,j])+np.float32(man_g[k][i+1,j+1])+np.float32(man_g[k][i+1,j+2])+
                                    np.float32(man_g[k][i+2,j])+np.float32(man_g[k][i+2,j+1])+np.float32(man_g[k][i+2,j+2]))/9))
 
```

## 4. Compute the standard deviation.

```python
man_std=list()
for i in range(16):
    man_std.append(np.std(man_g[i]))
```

## 5. Then we do normalization.

```python
man_n=[np.uint8(np.zeros((300,300)))for i in range(16)]
for i in range(16):
    man_n[i]=(man_g[i]-man_m[i])/man_std[i]
    
```

## "Plot the image"

```python
plt.imshow(man_m[0], cmap='Greys_r')
plt.show()
```

## 6. Then we convert the 2D matrix to 1D by using flatten() function.

```python
man_flat=[np.uint8(np.zeros((9000)))for i in range(16)]
for i in range(16):
    man_flat[i]=man_n[i].flatten()
```

## 7. We then compute the mean, kurtosis, variance and skewness of all the image.

```python
import scipy.stats as sc
man_var=list()
man_skew=list()
man_kur=list()
man_mean=list()
for i in range(16):
    man_mean.append(np.mean(man_flat[i]))
    man_var.append(np.var(man_flat[i]))
    man_skew.append(sc.skew(man_flat[i]))
    man_kur.append(sc.kurtosis(man_flat[i]))
    
man_mean=np.mean(man_mean)
man_var=np.mean(man_var)
man_skew=np.mean(man_skew)
man_kurtosis=np.mean(man_kur)
print("Mean       :"+str(man_mean))
print("Variance   :"+str(man_var))
print("Skewness   :"+str(man_skew))
print("Kurtosis   :"+str(man_kurtosis))
```
